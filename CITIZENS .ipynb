```python
def mapping_charges(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('importance'):
        m = subdf['rank'].median()
        sd = subdf['rank'].std()
        subd = subdf.copy()
        subd = subd[((subd['rank'] < m+5*sd))]
        reduced_df = subd
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return (df_out)
df=mapping_charges(df1)
```
